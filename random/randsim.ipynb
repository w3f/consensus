```python
import random
from enum import Enum
class slottype(Enum):
    GOOD=0
    BAD=1
    WEAK=2
def randslot():
    ran=random.random()
    if ran < goodprob:
        return slottype.GOOD
    if ran > 1- badprob:
        return slottype.BAD
    return slottype.WEAK
def generateepoch(epochlen):
    return [randslot() for slotnum in range(epochlen)]
epochlen = 600
slots=generateepoch(epochlen)
generateepoch(3)
```

```python
def maxadv(slots):
    advs=[]
    curradv=0
    for slot in slots:
        match slot:
            case slottype.GOOD:
                curradv -= 1
            case slottype.BAD:
                curradv += 1
        advs +=[curradv]
    return max(advs)
maxadvsim=0
dist=[0 for i in range(20)]
iters=10000
for i in range(iters):
    adv=maxadv(generateepoch(epochlen))
    if adv > maxadvsim:
        maxadvsim = adv
    if adv < 20 and adv >= 0:
        dist[adv] += RR(1/iters)
maxadvsim,dist      
```

```python
def randoptions(slots):
    negadvoptions=1
    options=[1 for i in range(15)]
    for slot in slots:
        match slot:
            case slottype.GOOD:
                for i in range(14,0,-1):
                    options[i]=options[i-1]
                options[0]=negadvoptions
                negadvoptions=1
            case slottype.BAD:
                for i in range(14,0,-1):
                    options[i]+=options[i-1]
                options[0]+=negadvoptions
                negadvoptions*=2
    return negadvoptions,options
randoptions(generateepoch(epochlen))
                
                    
```

```python
def epochssim(numepochs,epochlength):
    epoch=generateepoch(epochlen)
    epochoptions=[]
    for i in range(numepochs):
        lastepoch=epoch
        epoch=generateepoch(epochlen)
        negadvoptions,options=randoptions(lastepoch)
        adv=maxadv(epoch)
        if adv < 0:
            epochoptions +=[negadvoptions]
        elif adv > 14:
            print("high adv",adv)
        else:
            epochoptions +=[options[adv]]
    return epochoptions
a=epochssim(10000,600)
max(a),RR(sum(a)/1000),a   
```
